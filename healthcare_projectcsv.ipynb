```python

#How long patients each patients have stayed in the hospital?
data["Date of Admission"] = pd.to_datetime(data["Date of Admission"])
data["Discharge Date"] = pd.to_datetime(data["Discharge Date"])
data["Stay Duration"] = (data["Discharge Date"] - data["Date of Admission"])
stay_per_patient = data.groupby("Name")["Stay Duration"].sum().reset_index()
stay_per_patient.head(50)
```

How many patients are below 18, 18â€“40, 41â€“60, 60+?

```python
def patient_age(Age):
    if Age < 18:
        return "Child"
    elif 18 <= Age < 60:
        return "Adult"
    else:
        return "Senior"
data["Age Group"] = data["Age"].apply(patient_age)

data[["Age", "Age Group"]].head(50)
```

```python
data["Age Group"].value_counts().plot(kind='bar', color=['skyblue', 'orange', 'lightgreen'])
plt.title("Age Group Distribution")
plt.xlabel("Age Group")
plt.ylabel("Number of People")
plt.show()
```
